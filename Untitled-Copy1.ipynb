```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv('dados_sem_null_sql.csv')
dados
```

## Ordenando os dados por colunas

```python
dados.columns
```

```python
dados = dados[['id_cliente', 'id_emprestimo', 'id_historico', 'tipo_imovel', 
               'motivo_emprestimo', 'grupo', 'INADIMPLENCIA', 'idade', 'salario', 
               'tempo_trabalhando', 'valor_emprestimo', 'juros', 'emp_sal', 'tempo_cliente', 'RISCO_INADIMPLENCIA']]
dados
```

## TraduÃ§Ã£o dos dados que estÃ£o em inglÃªs

```python
tipo_prop= {'Rent': 'alugado', 
              'Own': 'proprio', 
              'Mortgage': 'financiado', 
              'Other': 'outro'}
motivo_emprestimo = {'Homeimprovement': 'reforma', 
              'Venture': 'risco', 
              'Personal': 'pessoal', 
              'Medical': 'saude', 
              'Debtconsolidation': 'quitar', 
              'Education': 'ensino'}
inadimplencia = {'N': 0, 'Y': 1}
```

```python
dados['tipo_imovel'] = dados.tipo_imovel.map(tipo_prop)
dados['motivo_emprestimo'] = dados.motivo_emprestimo.map(motivo_emprestimo)
dados['INADIMPLENCIA'] = dados.INADIMPLENCIA.map(inadimplencia)
dados.head()
```

## Apesar de remover os dados nulos no SQL, vamos verificar a existÃªncia de dados faltantes nos dados

```python
dados.isnull().sum()
```

## Como sÃ£o poucos dados em branco, podemos elimina-los.

```python
dados.dropna(inplace = True)
```

```python
dados.isnull().sum()
```

## Com os dados nulos removidos vamos para os 'outliers'. Vamos separar os dados numÃ©ricos e trata-los usando 'zscore'

```python
dados.columns
```

```python
dados_numericos = dados[['idade', 'salario', 'tempo_trabalhando', 
                    'valor_emprestimo', 'juros', 'emp_sal', 
                    'tempo_cliente', 'INADIMPLENCIA','RISCO_INADIMPLENCIA']]
dados_numericos.head()
```

```python
from scipy import stats
from scipy.stats import zscore
z_scores = stats.zscore(dados_numericos)
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
dados_sem_outliers = dados_numericos[filtered_entries]

dados_sem_outliers
```

## Agora vamos juntar os dados para manter as informaÃ§Ãµes dos nossos dados

```python
dados_categoricos = dados[['id_cliente', 'id_emprestimo', 
                           'id_historico', 'tipo_imovel',
                           'motivo_emprestimo', 'grupo']]
dados_categoricos.head()
```

```python
dados = dados_categoricos.merge(dados_sem_outliers, right_index = True, left_index=True)
dados
```

## Vamos iniciar com um classificador menos exigente para avaliar o comportamento dos dados em um modelo. Antes disso, vamos estabelecer nosso 'baseline' de classificaÃ§Ã£o com um Dummy Classifier.

```python
dados_numericos = dados[['idade', 'salario', 'tempo_trabalhando', 
                    'valor_emprestimo', 'juros', 'emp_sal', 
                    'tempo_cliente', 'INADIMPLENCIA','RISCO_INADIMPLENCIA']]
dados_numericos
```

```python
x = dados_numericos.drop(['RISCO_INADIMPLENCIA'], axis = 1)
y = dados_numericos['RISCO_INADIMPLENCIA']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.dummy import DummyClassifier
SEED = 301
np.random.seed(SEED)

modelo = DummyClassifier()
results = cross_validate(modelo, x, y, cv = 10, return_train_score=False)
media = results['test_score'].mean()
desvio_padrao = results['test_score'].std()
print("Accuracy com cross validation, 10 = [%.2f, %.2f]" % ((media - 2 * desvio_padrao)*100, (media + 2 * desvio_padrao) * 100))
```

```python
from sklearn.tree import DecisionTreeClassifier

modelo = DecisionTreeClassifier(max_depth=2)
results = cross_validate(modelo, x, y, cv = 3, return_train_score=False)
media = results['test_score'].mean()
desvio_padrao = results['test_score'].std()
print("Accuracy com cross validation, 10 = [%.2f, %.2f]" % ((media - 2 * desvio_padrao)*100, (media + 2 * desvio_padrao) * 100))
```

```python
from sklearn.model_selection import KFold

cv = KFold(n_splits = 10)
modelo = DecisionTreeClassifier(max_depth=2)
results = cross_validate(modelo, x, y, cv = cv, return_train_score=False)
media = results['test_score'].mean()
desvio_padrao = results['test_score'].std()
print("Accuracy com cross validation, 10 = [%.2f, %.2f]" % ((media - 2 * desvio_padrao)*100, (media + 2 * desvio_padrao) * 100))
```

```python
from sklearn.model_selection import KFold

cv = KFold(n_splits = 10, shuffle = True)
modelo = DecisionTreeClassifier(max_depth=2)
results = cross_validate(modelo, x, y, cv = cv, return_train_score=False)
media = results['test_score'].mean()
desvio_padrao = results['test_score'].std()
print("Accuracy com cross validation, 10 = [%.2f, %.2f]" % ((media - 2 * desvio_padrao)*100, (media + 2 * desvio_padrao) * 100))
```

```python
from sklearn.model_selection import GroupKFold
cv = GroupKFold(n_splits = 7)# o nÃºmero de splits deve corresponder ao nÃºmero de grupos, aqui vamos usar o grupo de emprestimos
modelo = DecisionTreeClassifier(max_depth=2)
modelo.fit(x, y)
results = cross_validate(modelo, x, y, cv = cv, groups = dados.grupo, return_train_score=False)
resultados = pd.DataFrame(data=results)
resultados
```

## Agora vamos usar o Grid Search para encontrar uma combinaÃ§Ã£o de hiperparÃ¢metros que melhores o nosso modelo.

```python
from sklearn.model_selection import GridSearchCV
lista_parametros = {
    "splitter" : ['best', 'random'],
    "max_depth" : [2, 3, 5],
    "min_samples_split" : [32, 64, 128],
    "min_samples_leaf" : [32, 64, 128],
    "criterion" : ['gini', 'entropy', 'log_loss']
}
busca = GridSearchCV(DecisionTreeClassifier(),
                    lista_parametros,
                    cv = GroupKFold(n_splits = 7))
busca.fit(x, y, groups = dados.grupo)
resultados = pd.DataFrame(busca.cv_results_)
resultados.head()
```

```python
print(busca.best_params_)
print(busca.best_score_ * 100)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.25,
                                                         stratify = y)
scaler = StandardScaler()
scaler.fit(treino_x)
treino_x_escalado = scaler.transform(treino_x)
teste_x_escalado = scaler.transform(teste_x)

modelo = DecisionTreeClassifier()
modelo.fit(treino_x_escalado, treino_y)
previsoes = modelo.predict(teste_x_escalado)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurÃ¡cia foi %.2f%%" % acuracia)
```

```python
dummy_stratified = DummyClassifier()
dummy_stratified.fit(treino_x, treino_y)
acuracia = dummy_stratified.score(teste_x, teste_y) * 100

print("A acurÃ¡cia do dummy stratified foi %.2f%%" % acuracia)
```

```python
SEED = 158020
np.random.seed(SEED)
modelo = DecisionTreeClassifier(max_depth=2)
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurÃ¡cia foi %.2f%%" % acuracia)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
pipe_knn = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=3), verbose=True)
pipe_knn.fit(treino_x, treino_y)
pipe_knn.score(treino_x, treino_y)
```

```python
pipe_tree = make_pipeline(StandardScaler(), DecisionTreeClassifier(max_depth=3), verbose=True)
pipe_tree.fit(treino_x, treino_y)
pipe_tree.score(treino_x, treino_y)
```

```python
from sklearn.naive_bayes import BernoulliNB
pipe_bnb = make_pipeline(StandardScaler(), BernoulliNB(), verbose=True)
pipe_bnb.fit(treino_x, treino_y)
pipe_bnb.score(treino_x, treino_y)
```

```python
from sklearn.svm import LinearSVC
pipe_svc = make_pipeline(StandardScaler(), LinearSVC(max_iter=10000), verbose=True)
pipe_svc.fit(treino_x, treino_y)
pipe_svc.score(treino_x, treino_y)
```

```python
from sklearn.tree import DecisionTreeClassifier
pipe_dtree = make_pipeline(StandardScaler(), DecisionTreeClassifier(), verbose=True)
pipe_dtree.fit(treino_x, treino_y)
pipe_dtree.score(treino_x, treino_y)
```

```python
from sklearn.model_selection import KFold

parameters = {'max_depth': [3, 4, 5]}
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
grid = GridSearchCV(pipe_dtree, param_grid=parameters, cv=kfold, n_jobs=-1)
grid.fit(treino_x, treino_y)
grid.best_params_
```
